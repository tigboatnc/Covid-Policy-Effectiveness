## 3.1 Cleaning up the state profile dataframe `CEN`

Basically removing __$__ , __\%__ , and __comma__

```python
censusDF = censusDF.replace(',','', regex=True)
censusDF = censusDF.replace('%','', regex=True)
censusDF = censusDF.replace('\$','', regex=True)
censusDF.head()
```

## 3.2 Cleaning Case peak dataset  `CP`
- Converting arrays to separate columns 
- Handling Missing Data so it dosent effect model

```python
case_peakDF.head()
```

```python
# Removing states with less than 3 peaks 
case_peakDF['peak_times'] =  case_peakDF["peak_dates"].apply(lambda x: len(x))
case_peakDF = case_peakDF[case_peakDF['peak_times']>2]
case_peakDF = case_peakDF.reset_index(drop=True)
```

```python
# Sorting the peak array 1st 2nd third peek
def sort_peak_array(money_str):
    money_str.sort()
    return money_str

case_peakDF['peak_dates'] = case_peakDF['peak_dates'].apply(sort_peak_array)
```

```python
split_df = pd.DataFrame(case_peakDF['peak_dates'].tolist())
case_peakDF_processed = pd.concat([case_peakDF, split_df], axis=1)
```

```python
case_peakDF_processed = case_peakDF_processed.add_prefix('CP_')
```

```python
case_peakDF_processed = case_peakDF_processed.merge(stateNameDF, left_on='CP_state', right_on='stateCode')
```

```python
COVID_START_DATE = datetime.strptime("2019-12-12", "%Y-%m-%d")

case_peakDF_processed['CP_0'] = pd.to_datetime(case_peakDF_processed['CP_0'])
case_peakDF_processed['CP_0_diff'] = case_peakDF_processed['CP_0'] - COVID_START_DATE
case_peakDF_processed['CP_1'] = pd.to_datetime(case_peakDF_processed['CP_1'])
case_peakDF_processed['CP_1_diff'] = case_peakDF_processed['CP_1'] - COVID_START_DATE
case_peakDF_processed['CP_2'] = pd.to_datetime(case_peakDF_processed['CP_2'])
case_peakDF_processed['CP_2_diff'] = case_peakDF_processed['CP_2'] - COVID_START_DATE
```

```python
case_peakDF_processed.head()
```

```python
# Converting number of days to integer 
def convert_days_to_int(days_timestamp):
    return int(days_timestamp.days)

case_peakDF_processed['CP_0_diff'] = case_peakDF_processed['CP_0_diff'].apply(convert_days_to_int)
case_peakDF_processed['CP_1_diff'] = case_peakDF_processed['CP_1_diff'].apply(convert_days_to_int)
case_peakDF_processed['CP_2_diff'] = case_peakDF_processed['CP_2_diff'].apply(convert_days_to_int)
```

```python
case_peakDF_processed = case_peakDF_processed.drop(columns=['stateCode',
                                                            'CP_peak_dates',
                                                            'CP_peak_properties',
                                                            'CP_peak_times',
                                                            'CP_Unnamed: 0',
                                                            'CP_0',
                                                           'CP_1',
                                                           'CP_2'])
```

## 3.6 Cleaning up the State Party Dataset `SP`

__Questionable decision here : Put NA as not available__ : WHY SON?  

#### Trivia and Domain Knowledge 
In some states the attorney general is marked as appointed, these are generally appointed by the governor so they are replaced there by governors affiliation

```python
print('Pre Cleanup')
statePartyDF.head()
```

```python
statePartyDF
statePartyDF = statePartyDF.fillna('NAHERE') # Not available Here
statePartyDF = statePartyDF[['Location',
                             'Governor Political Affiliation',
                             'State Senate Majority Political Affiliation',
                             'State House Majority Political Affiliation',
                            'State Attorney General Political Affiliation']]
```

```python
statePartyDF = statePartyDF.replace('Republican','0', regex=True)
statePartyDF = statePartyDF.replace('Democrat','1', regex=True)
statePartyDF = statePartyDF.replace('NAHERE','3', regex=True)
```

```python
statePartyDF.head()
```

```python
for index, row in statePartyDF.iterrows():
    if row['State Attorney General Political Affiliation'] == 'Appointed':
        statePartyDF.loc[index,'State Attorney General Political Affiliation'] = statePartyDF.loc[index,'Governor Political Affiliation']
        
```

```python
len_before = len(statePartyDF)
statePartyDF = statePartyDF.merge(stateNameDF, left_on='Location', right_on='stateName')

len_after = len(statePartyDF)

print(f'Length Before: {len_before}\nLength After: {len_after}\nState Length: {len(stateNameDF)}')
```

```python
statePartyDF.head()
```

```python
statePartyDF = statePartyDF.add_prefix('SP_')
```

## 4. Mergering Everything

Now since we have all the dataframes in order, we merge them to give us the training dataset

- statePartyDF
- hospitalBedDF
- fireDF
- obesityDF
- censusDF 
- case_peakDF_processed
- vacDF


```python
censusDF.head()
```

```python
TRAIN = statePartyDF.merge(hospitalBedDF, left_on='SP_stateCode', right_on='HB_stateCode')
```

```python
TRAIN = TRAIN.merge(fireDF, left_on='SP_stateCode', right_on='FD_stateCode')
```

```python
TRAIN = TRAIN.merge(obesityDF, left_on='SP_stateCode', right_on='OB_stateCode')
```

```python
TRAIN = TRAIN.merge(censusDF, left_on='SP_stateCode', right_on='CEN_stateCode')
```

```python
TRAIN = TRAIN.merge(case_peakDF_processed, left_on='SP_stateCode', right_on='CP_state')
```

```python
TRAIN = TRAIN.merge(vacDF, left_on='SP_stateName', right_on='S1_state')
```

```python
TRAIN.head()
```

# Cleaning up the final training data 

```python
# Dropping useless datasets
TRAIN = TRAIN.drop(columns=['SP_Location',
                    'SP_stateName',
                    'SP_stateCode',
                    'S1_state',
                   'HB_Location',
                    'HB_stateName',
                    'HB_stateCode',
                   'FD_stateCode',
                    'FD_stateName',
                    'OB_stateCode',
                    
                   'OB_stateName',
                           'CEN_stateName','CP_state'])
```

```python
old_columns = TRAIN.columns.to_list()
```

```python
old_columns.remove('stateName')
old_columns.remove('CEN_stateCode')
old_columns.sort()
```

```python
new_columns = ['stateName','CEN_stateCode']+ old_columns
```

```python
TRAIN = TRAIN[new_columns]
```
